```python
import urllib
import bs4
html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Hunter_%C3%97_Hunter')
bs = bs4.BeautifulSoup(html.read(),'html.parser')
print(bs.body.h1, end = "\n\n\n")
print(bs.img, end = "\n\n\n")
print(bs.a,end = "\n\n\n")
print(bs.ul,end = "\n\n\n")
print(bs.p)
print("\n\n\n\n")
print(bs.get_text)
```

```python
import urllib
import bs4
try:
    html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Hunter_%C3%97_Hunter')
except urllib.error.HTMLError as e:
    print('an html error has occured')
bs = bs4.BeautifulSoup(html.read(),'html.parser')
titles = bs.find_all(['h1','h2','h3','h4','h5','h6'])
for title in titles:
    print(title.get_text())
```

```python
import urllib
import bs4
try:
    html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Hunter_%C3%97_Hunter')
except urllib.error.HTMLError as e:
    print('an html error has occured')
bs = bs4.BeautifulSoup(html.read(),'html.parser')

tables = bs.findAll('table')
n = len(tables)
print(n)
print("\n\n")
for table in tables:
    print(table.get_text(), end = "\n\n\n\n")
```

```python
import urllib
import bs4
try:
    html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Hunter_%C3%97_Hunter')
except urllib.error.HTMLError as e:
    print('an html error has occured')
bs = bs4.BeautifulSoup(html.read(),'html.parser')

tables = bs.find('table',{'class':'infobox'})
print("\n\n")
for child in tables.children:
    print(child)
```

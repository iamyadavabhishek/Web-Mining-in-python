```python
import urllib
import bs4
import re

found = False

def graph(d,url,pageLeft,i):
    try:
        h = urllib.request.urlopen(url[d][i])
        b = bs4.BeautifulSoup(h.read(),'html.parser')
        nextBranch = []
        for links in b.findAll('a',href = re.compile(r'^(https:)')):
            if 'href' in links.attrs:
                nextBranch.append(links.attrs['href'])
        url.append(nextBranch)
        pageLeft = pageLeft - 1
        if(pageLeft>0):
            graph(d,url,pageLeft,i+1)
        elif(pageLeft == 0):
            return None
    except urllib.error.HTTPError as e:
        pageLeft = pageLeft - 1
        if(pageLeft>0):
            graph(d,url,pageLeft,i+1)
        elif(pageLeft == 0):
            return None
    except IndexError:
        return None

def findWord(url,word,n):
    for d in range(n,-1,-1):
        for i in range(len(url[d])):
            try:
                h = urllib.request.urlopen(url[d][i])
                b = bs4.BeautifulSoup(h.read(),'html.parser')
                content = b.find('body')
                print('searching: {0}'.format(url[d][i]))
                if word in content.get_text():
                    print("word found after traversing {0} pages in depth {1} on URL: {2}\n".format(i+1,d,(url[d][i]),))
                    return None
            except urllib.error.HTTPError as e:
                continue
        print('word not found in depth {0}'.format(d))
            
       
if __name__ == "__main__":
    urlMain = input('Enter the website you want to search for: ')
    url = []
    url.append([urlMain])
    word = input('enter the word you want to search: ')
    n = int(input('enter the depth you want to search the internet: '))
    for d in range(n):
        i = 0
        graph(d,url,len(url[d]),i)
    
    findWord(url,word,n)
```
